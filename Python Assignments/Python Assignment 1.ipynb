```python
import requests
from lxml import html
import re

# Define the URL of the webpage you want to scrape
url = 'https://www.unitedsol.net/list-of-top-it-companies-in-pakistan/'

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    
    # Parse the HTML content of the page with lxml
    tree = html.fromstring(response.content)

    # Use CSS selectors to find elements
    elements = tree.cssselect('p, h1, h2, h3')  # Example: all paragraphs and headers

    # Extract and clean text from selected elements
    text_list = [element.text_content() for element in elements]
    clean_text = ' '.join(text_list)
    clean_text = re.sub(r'\s+', ' ', clean_text).strip()

    # Print the cleaned text
    print(clean_text)

else: 
    print("Try Again and Again")
```
